```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time 
```

## scrap first page 

```python
url = page_links[0]
response = requests.get(url)
soup = BeautifulSoup(response.text , 'html.parser')
```

```python
soup
```

```python
all_jobs_1 = []
```

```python
job_cards = soup.find_all('div',{'class':'css-ghe2tq e1v1l3u10'})
```

```python
for job in job_cards:
  # for job title
  title_tag = job.find("h2").find("a")
  title = title_tag.text.strip()
  link = title_tag["href"]


  #for company
  company_tag = job.find("a", class_="css-ipsyv7")
  company = company_tag.text.replace("-", "").strip() if company_tag else "Confidential"


  #location 
  location_tag = job.find("span", class_="css-16x61xq")
  location = location_tag.text.strip() if location_tag else None

  # Posted Time
  time_tag = job.find("div", class_="css-eg55jf")
  posted = time_tag.text.strip() if time_tag else None

  # Job Type & Work Mode
  badges = job.find_all("span", class_="css-uc9rga") + job.find_all("span", class_="css-uofntu")
  badges_text = [b.text.strip() for b in badges]

  # Experience & Category Line
  exp_line = job.find("div", class_="css-1rhj4yg").text.strip()

  #  Skills (bottom tags)
  skills = [s.text.strip() for s in job.find_all("a", class_="css-5x9pm1")]

  all_jobs_1.append(
  {
    "title": title,
            "company": company,
            "location": location,
            "posted": posted,
            "badges": ", ".join(badges_text),
            "experience_line": exp_line,
            "skills": ", ".join(skills),
            "link": link
  }
)




```

```python
df = pd.DataFrame(all_jobs_1)
```

# scrap all the website

```python
headers = {"User-Agent": "Mozilla/5.0"}
all_jobs = []

page_links = [f'https://wuzzuf.net/search/jobs/?a=hpb&q=&start={i}' for i in range(655)]
```

```python
# scrap all the web site

for url in page_links:
    print("Scraping:", url)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    job_cards = soup.find_all('div', {'class': 'css-ghe2tq e1v1l3u10'})

    for job in job_cards:
        title_tag = job.find("h2").find("a")
        title = title_tag.text.strip()
        link = "https://wuzzuf.net" + title_tag["href"]

        company_tag = job.find("a", class_="css-ipsyv7")
        company = company_tag.text.replace("-", "").strip() if company_tag else "Confidential"

        location_tag = job.find("span", class_="css-16x61xq")
        location = location_tag.text.strip() if location_tag else None

        time_tag = job.find("div", class_="css-eg55jf")
        posted = time_tag.text.strip() if time_tag else None

        badges = job.find_all("span", class_="css-uc9rga") + job.find_all("span", class_="css-uofntu")
        badges_text = [b.text.strip() for b in badges]

        exp_line_tag = job.find("div", class_="css-1rhj4yg")
        exp_line = exp_line_tag.text.strip() if exp_line_tag else None

        skills = [s.text.strip() for s in job.find_all("a", class_="css-5x9pm1")]

        all_jobs.append({
            "title": title,
            "company": company,
            "location": location,
            "posted": posted,
            "badges": ", ".join(badges_text),
            "experience_line": exp_line,
            "skills": ", ".join(skills),
            "link": link
        })

    time.sleep(2)  # polite delay

df = pd.DataFrame(all_jobs)
print(df.shape)
```
